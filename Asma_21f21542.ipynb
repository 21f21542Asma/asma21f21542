```python
import scipy.stats as stats
import pandas as pd
import numpy as np

# Assuming dataTrain is your DataFrame

# 1. Select only numerical columns for z-score calculation
numerical_data = dataTrain.select_dtypes(include=[np.number])

# 2. Calculate z-scores for the numerical data
dataTrain_zscore = stats.zscore(numerical_data)

# 3. Convert the result back to a DataFrame (if needed)
dataTrain_zscore = pd.DataFrame(dataTrain_zscore, index=numerical_data.index, columns=numerical_data.columns)

# 4. If you need to keep the non-numerical columns, merge back with the original DataFrame
dataTrain = dataTrain.drop(columns=numerical_data.columns, errors='ignore')  # Remove the original numerical columns
dataTrain = pd.concat([dataTrain, dataTrain_zscore], axis=1) # Concatenate the z-score columns
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import category_encoders as ce

# Assuming dataTrain is your DataFrame

# 1. Check the actual column names in your DataFrame
print(dataTrain.columns)

# 2. Adjust the 'cols' parameter of the OneHotEncoder to match the actual column names
# For example, if your DataFrame has 'street' instead of 'address', change it accordingly:
encoder = ce.OneHotEncoder(cols=['street', 'country', 'city', 'statezip'],
                           handle_unknown='ignore', use_cat_names=True)

# Fit the encoder on the training data and transform both training and testing data
dataTrain_encoded = encoder.fit_transform(dataTrain)

# Split the encoded data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
   dataTrain_encoded.drop('price', axis=1),  # Features (excluding 'price')
   dataTrain_encoded['price'],                # Target variable
   test_size=0.2,
   random_state=42
)

# Create and train the Linear Regression model
model = LinearRegression()
model_mlr = model.fit(x_train, y_train)
```
